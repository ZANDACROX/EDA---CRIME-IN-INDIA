# **1. CRIME RATE PER LAKH -YEARS 2022 & 2021 - IPC WISE AND SLL WISE**

in this , we will see that crime rate per lakh for the years 2022 and 2021, both IPC and SLL wise.

2.   ***CRIME RATE SLL.csv***



These files can be found in the github repo: https://github.com/ZANDACROX/EDA---CRIME-IN-INDIA

```python
#from google.colab import files
#uploaded = files.upload()
import pandas as pd
import matplotlib.pyplot as plt

import pandas as pd

# Replace with the raw URL of the CSV file
url1= "https://raw.githubusercontent.com/ZANDACROX/EDA---CRIME-IN-INDIA/refs/heads/main/CRIME%20RATE%20IPC.csv"
url2= 'https://raw.githubusercontent.com/ZANDACROX/EDA---CRIME-IN-INDIA/refs/heads/main/CRIME%20RATE%20SLL.csv'
df = pd.read_csv(url1)
df2 = pd.read_csv(url2)
print("CRIME RATE PER LAKH-IPC\n")
print(df.to_string())
print("\n\n\n\nCRIME RATE PER LAKH-SLL\n")

print(df2.to_string())
```

FOR 2022 : IPC

```python
from bokeh.models import GeoJSONDataSource, HoverTool, LinearColorMapper, ColorBar, FixedTicker
from bokeh.palettes import Reds9  # Using 9 shades of red
from bokeh.plotting import figure
from bokeh.io import output_file, show, output_notebook
from urllib.request import urlopen
import json
import pandas as pd
print('\n\n\nCRIME RATE PER LAKH OF ALL THE STATES IN 2022 - IPC\n\n\n\n\n')
# Display Bokeh output in notebook
output_notebook()

# Load the Excel data and process it

# Filter for the year 2022
df_2022 = df[df['Year'] == 2022].copy()

# Define crime rate classification function
def classify_crime_rate(rate):
    if rate < 50:
        return 0  # Lightest shade of red
    elif rate < 100:
        return 1
    elif rate < 150:
        return 2
    elif rate < 200:
        return 3
    elif rate < 250:
        return 4
    elif rate < 300:
        return 5
    elif rate < 350:
        return 6
    elif rate < 400:
        return 7
    else:
        return 8  # Darkest shade of red

# Apply the classification to create a color class column
df_2022['Color_Class'] = df_2022['Crime Rate / Lakh'].apply(classify_crime_rate)

# Define a mapping for state name exceptions
state_name_mapping = {
    'Odisha': 'Orissa',
    'Jammu & Kashmir': 'Jammu and Kashmir',
    'A&N Islands': 'Andaman and Nicobar'
}

# Standardize state names in the DataFrame
df_2022['State/UT'] = df_2022['State/UT'].replace(state_name_mapping)

# Load GeoJSON data for India
with urlopen("https://raw.githubusercontent.com/geohacker/india/master/state/india_state.geojson") as response:
    geojson = json.load(response)

# Map each state in the GeoJSON data to the crime rate class and actual rate from df_2022
state_to_color_class = df_2022.set_index('State/UT')['Color_Class'].to_dict()
state_to_crime_rate = df_2022.set_index('State/UT')['Crime Rate / Lakh'].to_dict()

# Reverse the Reds9 palette to match classification order
color_list = Reds9[::-1]  # Reversed Reds9 palette

# Assign the appropriate shade of red and crime rate to each state
for feature in geojson['features']:
    state_name = feature['properties']['NAME_1']
    color_class = state_to_color_class.get(state_name, 0)  # Default to lightest shade if not found
    crime_rate = state_to_crime_rate.get(state_name, "N/A")  # Default to "N/A" if not found
    feature['properties']['Color'] = color_list[color_class]
    feature['properties']['CrimeRate'] = crime_rate

# Configure GeoJSON data source
geo_source = GeoJSONDataSource(geojson=json.dumps(geojson))

# Configure color mapper with reversed palette
cmap = LinearColorMapper(palette=color_list, low=0, high=8)

# Define custom tick labels for each crime rate range
color_bar = ColorBar(color_mapper=cmap,
                     ticker=FixedTicker(ticks=list(range(9))),  # Set tick positions for each color
                     location=(0, 0),
                     title='Crime Rate / Lakh',
                     major_label_overrides={
                         0: '0-50',
                         1: '50-100',
                         2: '100-150',
                         3: '150-200',
                         4: '200-250',
                         5: '250-300',
                         6: '300-350',
                         7: '350-400',
                         8: '400+'
                     })

# Set up the plot
TOOLS = "pan,wheel_zoom,box_zoom,reset,hover,save"
p = figure(title='India Crime Rate per Lakh (2022) - IPC', tools=TOOLS,
           x_axis_location=None, y_axis_location=None, width=800, height=800)
p.grid.grid_line_color = None

# Plot patches with the assigned color
p.patches('xs', 'ys', fill_alpha=0.7, line_color='black',
          fill_color='Color', line_width=0.1, source=geo_source)

# Configure hover tool to show the state name and crime rate
hover = p.select_one(HoverTool)
hover.point_policy = 'follow_mouse'
hover.tooltips = [('State:', '@NAME_1'), ('Crime Rate / Lakh:', '@CrimeRate')]

# Add color bar (legend) to the plot
p.add_layout(color_bar, 'right')

# Display the plot
show(p)



```
