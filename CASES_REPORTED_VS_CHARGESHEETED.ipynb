# **1. YEAR WISE NUMBER OF CASES REPORTED VERSUS NUMBER OF CASES CHARGE SHEETED BY CRIME-HEAD- 5 IPC & 5 SLL**

First, we will take 5 IPC crimes and 5 SLL crimes and see the number of cases reported vs charge sheeted for each year for 5 years (2018-2022)

The 5 IPC crimes are:


*   MURDER
*   RAPE
*   KIDNAPPING AND ABDUCTION
*   EXTORSION AND BLACKMAILING
*   DACOITY

The 5 SLL crimes are:


*   THE DOWRY PROHIBITION ACT
*   THE PROTECTION OF CHILDREN FROM SEXUAL OFFENCES ACT
*   THE PASSPORT ACT
*   THE CHILD LABOUR ACT
*   PROHIBITION ACT (STATE)


Uploading the respective csv file: crimehead wise reported vs chargesheeted ipc sll.csv

This file can be found in tthe GitHub repo.

```python
#from google.colab import files
#uploaded = files.upload()

import pandas as pd

# Replace with the raw URL of the CSV file
url = 'https://raw.githubusercontent.com/ZANDACROX/EDA---CRIME-IN-INDIA/refs/heads/main/crimehead%20wise%20reported%20vs%20chargesheeted%20ipc%20sll.csv'
df = pd.read_csv(url)

df
```

Now we will into these cases year-wise.

2018:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# IPC Data Processing and Plotting
ipc = ['Murder', 'Rape', 'Kidnapping and Abduction', 'Extortion and Blackmailing', 'Dacoity']
sll = ['The Dowry Prohibition Act', 'The Protection of Children from Sexual Offences Act', 'The Passport Act','The Child Labour (Prohibition & Regulation) Act', 'Prohibition Act (State)']

dataipc = []
for i in ipc:
    filtered_df = df[(df['Year'] == 2018) & (df['Crime Head'] == i) & (df['Crime Type'] == 'IPC')]
    a1 = filtered_df['Cases Reported during year'].apply(pd.to_numeric, errors='coerce').dropna().tolist()
    a2 = filtered_df['Charge Sheeted'].apply(pd.to_numeric, errors='coerce').dropna().tolist()
    dataipc.extend(a1)
    dataipc.extend(a2)

plt.figure(figsize=(8, 6))
x = np.arange(len(dataipc) + len(dataipc) // 2)
adjusted_bar_values = []
for i, value in enumerate(dataipc):
    adjusted_bar_values.append(value)
    if (i + 1) % 2 == 0:
        adjusted_bar_values.append(np.nan)

bar_color = 'skyblue'
plt.bar(x, adjusted_bar_values, color=[bar_color if not np.isnan(v) else 'white' for v in adjusted_bar_values])

labels = []
k = 0
bar_label = ipc[k] + ' Reported'
for i in range(len(adjusted_bar_values)):
    if np.isnan(adjusted_bar_values[i]):
        labels.append('')
    else:
        labels.append(bar_label)
        if bar_label == (ipc[k] + ' Reported'):
            bar_label = ipc[k] + ' Chargesheeted'
        else:
            k += 1
            if k < len(ipc):
                bar_label = ipc[k] + ' Reported'

plt.xlabel("IPC CRIMES REPORTED AND CHARGESHEETED")
plt.ylabel("NUMBER OF CASES")
plt.title("No. OF IPC CRIMES REPORTED AND CHARGESHEETED IN 2018")
plt.xticks(ticks=x, labels=labels, rotation=90)
print("FOR 5 IPC CRIMES:\n")
plt.show()

# SLL Data Processing and Plotting
datasll = []
for i in sll:
    filtered_df = df[(df['Year'] == 2018) & (df['Crime Head'] == i) & (df['Crime Type'] == 'SLL')]
    a1 = filtered_df['Cases Reported during year'].apply(pd.to_numeric, errors='coerce').dropna().tolist()
    a2 = filtered_df['Charge Sheeted'].apply(pd.to_numeric, errors='coerce').dropna().tolist()
    datasll.extend(a1)
    datasll.extend(a2)

plt.figure(figsize=(8, 6))
x = np.arange(len(datasll) + len(datasll) // 2)
adjusted_bar_values = []
for i, value in enumerate(datasll):
    adjusted_bar_values.append(value)
    if (i + 1) % 2 == 0:
        adjusted_bar_values.append(np.nan)

bar_color = 'skyblue'
plt.bar(x, adjusted_bar_values, color=[bar_color if not np.isnan(v) else 'white' for v in adjusted_bar_values])

labels = []
k = 0
bar_label = sll[k] + ' Reported'
for i in range(len(adjusted_bar_values)):
    if np.isnan(adjusted_bar_values[i]):
        labels.append('')
    else:
        labels.append(bar_label)
        if bar_label == (sll[k] + ' Reported'):
            bar_label = sll[k] + ' Chargesheeted'
        else:
            k += 1
            if k < len(sll):
                bar_label = sll[k] + ' Reported'

plt.xlabel("SLL CRIMES REPORTED AND CHARGESHEETED")
plt.ylabel("NUMBER OF CASES")
plt.title("No. OF SLL CRIMES REPORTED AND CHARGESHEETED IN 2018")
plt.xticks(ticks=x, labels=labels, rotation=90)
print("\n\nFOR 5 SLL CRIMES:\n")
plt.show()
```
